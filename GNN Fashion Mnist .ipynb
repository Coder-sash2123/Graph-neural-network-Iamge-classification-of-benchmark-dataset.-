```python
import torch
from torch import nn
import torch.nn.functional as F
from torch_geometric.nn import GCNConv, global_mean_pool
from torch_geometric.data import Data, DataLoader
from torchvision import datasets, transforms
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# 1. Load Fashion MNIST dataset and convert images to graph data
def create_graph_from_image(image):
    # Convert the image tensor to a numpy array
    image_np = image.numpy().squeeze()
    rows, cols = image_np.shape
    
    # Prepare node features
    nodes = []
    edges = []
    node_features = []

    for i in range(rows):
        for j in range(cols):
            node_index = i * cols + j
            nodes.append(node_index)
            node_features.append([image_np[i, j]])

            # Add edges between the current pixel and its 4-connected neighbors
            if i > 0:  # Top
                edges.append((node_index, (i - 1) * cols + j))
            if i < rows - 1:  # Bottom
                edges.append((node_index, (i + 1) * cols + j))
            if j > 0:  # Left
                edges.append((node_index, i * cols + (j - 1)))
            if j < cols - 1:  # Right
                edges.append((node_index, i * cols + (j + 1)))

    # Create edge index from edges
    edge_index = torch.tensor(edges, dtype=torch.long).t().contiguous()
    x = torch.tensor(node_features, dtype=torch.float)
    return Data(x=x, edge_index=edge_index)

# 2. Load FashionMNIST and preprocess data
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Resize((8, 8)),  # Resize to 8x8 to reduce graph size
    transforms.Normalize((0.5,), (0.5,))  # Normalize to [0,1] range
])

fashion_mnist_train = datasets.FashionMNIST(root='./data', train=True, transform=transform, download=True)
fashion_mnist_test = datasets.FashionMNIST(root='./data', train=False, transform=transform, download=True)

# 3. Convert FashionMNIST images to graphs
train_graphs = [create_graph_from_image(image) for image, label in fashion_mnist_train]
for graph, (image, label) in zip(train_graphs, fashion_mnist_train):
    graph.y = torch.tensor([label], dtype=torch.long)

test_graphs = [create_graph_from_image(image) for image, label in fashion_mnist_test]
for graph, (image, label) in zip(test_graphs, fashion_mnist_test):
    graph.y = torch.tensor([label], dtype=torch.long)

train_loader = DataLoader(train_graphs, batch_size=64, shuffle=True)
test_loader = DataLoader(test_graphs, batch_size=64, shuffle=False)

# 4. Build a GNN model using torch_geometric for classification
class GNN(torch.nn.Module):
    def __init__(self):
        super(GNN, self).__init__()
        self.conv1 = GCNConv(1, 16)
        self.conv2 = GCNConv(16, 32)
        self.fc1 = nn.Linear(32, 10)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = self.conv2(x, edge_index)
        x = F.relu(x)
        x = global_mean_pool(x, data.batch)  # Correctly reference global_mean_pool here
        x = self.fc1(x)
        return F.log_softmax(x, dim=1)

# 5. Train and evaluate the GNN model
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = GNN().to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
criterion = nn.CrossEntropyLoss()

def train():
    model.train()
    total_loss = 0
    correct = 0
    for data in train_loader:
        data = data.to(device)
        optimizer.zero_grad()
        out = model(data)
        loss = criterion(out, data.y)
        loss.backward()
        optimizer.step()
        total_loss += loss.item() * data.num_graphs
        pred = out.argmax(dim=1)
        correct += pred.eq(data.y).sum().item()
    return total_loss / len(train_loader.dataset), correct / len(train_loader.dataset)

def test(loader):
    model.eval()
    total_loss = 0
    correct = 0
    with torch.no_grad():
        for data in loader:
            data = data.to(device)
            out = model(data)
            loss = criterion(out, data.y)
            total_loss += loss.item() * data.num_graphs
            pred = out.argmax(dim=1)
            correct += pred.eq(data.y).sum().item()
    return total_loss / len(loader.dataset), correct / len(loader.dataset)

# Training loop
epochs = 350
for epoch in range(epochs):
    train_loss, train_acc = train()
    test_loss, test_acc = test(test_loader)
    print(f"Epoch {epoch + 1}, Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f}, "
          f"Test Loss: {test_loss:.4f}, Test Acc: {test_acc:.4f}")

# 6. Visualize one of the images as a graph
def visualize_graph(data):
    G = nx.Graph()
    
    # Add nodes and edges to the graph
    for i, feature in enumerate(data.x):
        G.add_node(i, value=feature.item())
    edge_index = data.edge_index.t().numpy()
    for edge in edge_index:
        G.add_edge(edge[0], edge[1])
    
    # Create a new figure and axes
    fig, ax = plt.subplots(figsize=(8, 8))
    
    # Generate the layout for the graph
    pos = nx.spring_layout(G, seed=42)  # Fixed random seed for consistent layout
    
    # Draw the graph
    node_colors = [data.x[i].item() for i in range(len(data.x))]  # Use pixel intensity as node color
    nx.draw(G, pos, node_size=50, node_color=node_colors, cmap='coolwarm', with_labels=False, ax=ax)
    
    # Show the plot
    plt.show()

# Visualize the first graph in the test set
visualize_graph(test_graphs[0])



```

```python
import os
import torch
import numpy as np
import torch.nn.functional as F
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from torch_geometric.data import Data, DataLoader
from skimage import io, segmentation
import networkx as nx
from torch_geometric.nn import GCNConv, global_mean_pool

# Function to create graph from image
def create_graph_from_image(image, superpixel_labels):
    # Get the number of superpixels
    num_superpixels = np.max(superpixel_labels) + 1

    # Create the node features as the mean color of each superpixel
    node_features = []
    for i in range(num_superpixels):
        mask = superpixel_labels == i
        node_features.append(np.mean(image[mask], axis=0))  # Mean color of the superpixel
    node_features = np.array(node_features)

    # Create an empty graph
    edge_index = []
    for i in range(num_superpixels):
        for j in range(i + 1, num_superpixels):
            if np.any(superpixel_labels == i) and np.any(superpixel_labels == j):
                # Add edge if the superpixels are neighbors
                edge_index.append([i, j])
                edge_index.append([j, i])
    edge_index = torch.tensor(edge_index, dtype=torch.long).t().contiguous()

    # Convert node features into tensor
    x = torch.tensor(node_features, dtype=torch.float)

    # Create a Data object
    data = Data(x=x, edge_index=edge_index)
    return data

# Load the data from the image directory
def load_data(image_dir):
    graphs = []
    labels = []

    # Check the files in the directory
    print(f"Files in {image_dir}: {os.listdir(image_dir)}")  # Debugging line

    # Iterate over all files in the directory
    for filename in os.listdir(image_dir):
        if filename.endswith(".png") or filename.endswith(".jpg") or filename.endswith(".jpeg"):  # Add more formats if needed
            image_path = os.path.join(image_dir, filename)
            print(f"Loading image: {filename}")  # Debugging line
            image = io.imread(image_path)
            
            # Check if image is loaded correctly
            if image is None:
                print(f"Failed to load image: {filename}")
                continue
                
            label = int(filename.split('_')[0])  # Adjust this if labels are elsewhere

            # Perform superpixel segmentation
            superpixel_labels = segmentation.slic(image, n_segments=10, compactness=10, sigma=1)

            # Create the graph for the image
            graph = create_graph_from_image(image, superpixel_labels)
            graph.y = torch.tensor([label])  # Assign label to the graph

            graphs.append(graph)
            labels.append(label)

    if not graphs:
        print("No graphs were created. Please check the image directory and filenames.")

    return graphs

# Define the GNN model
class GNN(torch.nn.Module):
    def __init__(self, num_features, num_classes):
        super(GNN, self).__init__()
        self.conv1 = GCNConv(num_features, 64)
        self.conv2 = GCNConv(64, 32)
        self.fc = torch.nn.Linear(32, num_classes)

    def forward(self, data):
        x, edge_index, batch = data.x, data.edge_index, data.batch
        x = F.relu(self.conv1(x, edge_index))
        x = F.relu(self.conv2(x, edge_index))
        x = global_mean_pool(x, batch)
        x = self.fc(x)
        return x

# Function to train the model
def train(loader, model, optimizer, criterion):
    model.train()
    total_loss = 0
    correct = 0
    total = 0
    for data in loader:
        optimizer.zero_grad()
        data = data.to(device)
        out = model(data)
        loss = criterion(out, data.y)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        pred = out.argmax(dim=1)
        correct += (pred == data.y).sum().item()
        total += data.y.size(0)

    return total_loss / len(loader), correct / total

# Function to test the model
def test(loader, model, criterion):
    model.eval()
    total_loss = 0
    correct = 0
    total = 0
    with torch.no_grad():
        for data in loader:
            data = data.to(device)
            out = model(data)
            loss = criterion(out, data.y)
            total_loss += loss.item()
            pred = out.argmax(dim=1)
            correct += (pred == data.y).sum().item()
            total += data.y.size(0)

    return total_loss / len(loader), correct / total

# Visualize graph
def visualize_graph(data):
    G = nx.Graph()
    pos = {}

    # Add nodes and edges to the graph
    for i in range(data.num_nodes):
        G.add_node(i, feature=data.x[i].numpy())
    
    for edge in data.edge_index.t().tolist():
        G.add_edge(edge[0], edge[1])
    
    # Visualize the graph
    node_colors = [data.x[i].item() for i in range(len(data.x))]
    plt.figure(figsize=(8, 8))
    nx.draw(G, pos, node_size=50, node_color=node_colors, cmap='coolwarm', with_labels=False)
    plt.show()

# Main code to train and evaluate the GNN model
image_dir = "./data/FashionMNIST/raw"  # Specify the correct path
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Load the data
graphs = load_data(image_dir)

# Split the data into training and test sets
train_graphs, test_graphs = train_test_split(graphs, test_size=0.2, random_state=42)

# Create DataLoader for batching
train_loader = DataLoader(train_graphs, batch_size=32, shuffle=True)
test_loader = DataLoader(test_graphs, batch_size=32, shuffle=False)

# Initialize model, optimizer, and loss function
num_features = train_graphs[0].x.shape[1]
num_classes = len(set([graph.y.item() for graph in train_graphs]))  # Number of unique labels
model = GNN(num_features, num_classes).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
criterion = torch.nn.CrossEntropyLoss()

# Train and test the model
epochs = 10
for epoch in range(epochs):
    train_loss, train_acc = train(train_loader, model, optimizer, criterion)
    test_loss, test_acc = test(test_loader, model, criterion)
    print(f'Epoch {epoch + 1} | Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f} | Test Loss: {test_loss:.4f} | Test Acc: {test_acc:.4f}')

# Visualize the first graph in the test set
visualize_graph(test_graphs[0])

```
